```python
import pandas as pd
```

```python
playersdf = pd.read_csv('players.csv')
df_2019 = pd.read_csv('ratings_2019.csv')
players_2019 = playersdf.merge(df_2019, on='fide_id')
```

```python
players_2019.head()
```

```python
del players_2019['rating_rapid']
del players_2019['rating_blitz']
players_2019.head()
```

```python
playersdf = pd.read_csv('players.csv')
df_2020 = pd.read_csv('ratings_2020.csv')
players_2020 = playersdf.merge(df_2020, on='fide_id')
```

```python
del players_2020['rating_rapid']
del players_2020['rating_blitz']
del players_2020['month']
players_2020.head()
```

```python
players_2019_2020 = players_2019.merge(players_2020, on='fide_id')
players_2019_2020.head(50)
```

```python
del players_2019_2020['name_y']
del players_2019_2020['federation_y']
del players_2019_2020['gender_y']
del players_2019_2020['yob_y']
```

```python
players_2019_2020.head(20)
```

```python
cleaned_players=players_2019_2020.drop_duplicates(subset=['fide_id', 'month'], keep='first', inplace=False)
```

```python
cleaned_players.head(50)
```

```python
players_2019.to_csv(r'players_2019.csv', index = False)
```

```python
players_2020.to_csv(r'players_2020.csv', index = False)
```

```python
cleaned_players.to_csv(r'players_2019_2020.csv', index = False)
```
