```python
import pandas as pd
import numpy as np
```

```python
df_2020 = pd.read_csv('players_2020.csv', dtype='unicode')
```

```python
df_2020.head()
```

```python
sorted_df = df_2020.sort_values(['federation','rating_standard'] , ascending=[True, False])
sorted_df.head(50)
```

```python
drop_df = sorted_df.drop_duplicates(subset='federation', keep='first')
drop_df.head(50)
```

```python
sorted_df['rating_standard'] = pd.to_numeric(sorted_df['rating_standard'])
```

```python
sorted_df.dtypes
```

```python
average_df = sorted_df.groupby('federation').rating_standard.apply(np.mean)
average_df.head()
```

```python
top_50 = average_df.sort_values(ascending=False)
top_50.head(50)
```

```python
sorted_name = df_2020.sort_values(['name','rating_standard'], ascending=(True, False))
sorted_name.head(50)
```

```python
single_name = sorted_name.drop_duplicates(subset='name', keep='first')
single_name.head(50)
```

```python
sorted_single_name = single_name.sort_values(['federation','rating_standard'], ascending=(True, False))
sorted_single_name.head(50)
```

```python
number_of_players = sorted_single_name['federation'].value_counts()
number_of_players
```
